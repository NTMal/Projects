<div id="dataset_analysis">
    <h2>Dataset Analysis</h2>
</div>


- Each row in the dataset represents a credit card transaction. As shown above, each row has 31 variables. One variable (the last variable in the table above) is called Class and represents the target variable. Train model using the other variables to predict the value of the Class variable. Let's first retrieve basic statistics about the target variable.

- For confidentiality reasons, the original names of most features are anonymized V1, V2 .. V28. The values of these features are the result of a PCA transformation and are numerical. The feature 'Class' is the target variable and it takes two values: 1 in case of fraud and 0 otherwise. For more information about the dataset please visit this webpage: https://www.kaggle.com/mlg-ulb/creditcardfraud.

```python
# Get the set of distinct classes
labels = raw_data.Class.unique()

# Get the count of each class
sizes = raw_data.Class.value_counts().values

# Plot the class value counts
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.3f%%')
ax.set_title('Target Variable Value Counts')
plt.show()
```

- As shown above, the Class variable has two values: 0 (the credit card transaction is legitimate) and 1 (the credit card transaction is fraudulent). Thus, need to model a binary classification problem.
- The dataset is highly unbalanced, the target variable classes are not represented equally. This case requires special attention when training or when evaluating the quality of a model. One way of handing this case at train time is to bias the model to pay more attention to the samples in the minority class. The models under the current study will be configured to take into account the class weights of the samples at train/fit time.


It is also prudent to understand which features affect the model in what way. Visualise the effect of the different features on the model by computing the features correlation.


```python
correlation_values = raw_data.corr()['Class'].drop('Class')
correlation_values.plot(kind='barh', figsize=(10, 6))
```

This clearly shows that some features affect the output Class more than the others. For efficient modeling, we may use only the most correlated features.


<div id="dataset_preprocessing">
    <h2>Dataset Preprocessing</h2>
</div>


- Prepare the data for training.
- Apply standard scaling to the input features and normalize them using $L_1$ norm for the training models to converge quickly.
- As seen in the data snapshot, there is a parameter called `Time` which will not be considered for modeling because time of transaction occurs doesn't tell whether it is fraudulent. Fraud can happen any time of the day.
- Hence, features 2 to 30 will be used as input features and feature 31, i.e. Class will be used as the target variable.

```python
# Standardise features by removing the mean and scaling to unit variance
raw_data.iloc[:, 1:30] = StandardScaler().fit_transform(raw_data.iloc[:, 1:30])
data_matrix = raw_data.values

# X: feature matrix (for this analysis, we exclude the Time variable from the dataset)
X = data_matrix[:, 1:30]

# y: labels vector
y = data_matrix[:, 30]

# Data normalisation
X = normalize(X, norm="l1")
```

<div id="dataset_split">
    <h2>Dataset Train/Test Split</h2>
</div>


- Now the dataset is ready for building the classification models.
- First divide the pre-processed dataset into a subset to be used for training the model (the train set) and a subset to be used for evaluating the quality of the model (the test set).

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

<div id="dt_sklearn">
    <h2>Build a Decision Tree Classifier model with Scikit-Learn</h2>
</div>


Compute the sample weights to be used as input to the train routine so that it takes into account the class imbalance present in this dataset.


```python
w_train = compute_sample_weight('balanced', y_train)
```

Using these sample weights, train the Decision Tree classifier. Make note of the time it takes for training this model to compare it against SVM later.

```python
# For reproducible output across multiple function calls, set random_state to a given integer value
dt_all = DecisionTreeClassifier(max_depth=4, random_state=42)

dt_all.fit(X_train, y_train, sample_weight=w_train)
```

<div id="svm_sklearn">
    <h2>Build a Support Vector Machine model with Scikit-Learn</h2>
</div>


Unlike Decision Trees, no need to initiate a separate sample_weight for SVMs. Simply pass a parameter in the scikit-learn function.


```python
# For reproducible output across multiple function calls, set random_state to a given integer value
svm_all = LinearSVC(class_weight='balanced', 
                random_state=42, loss="hinge", fit_intercept=False)

svm_all.fit(X_train, y_train)
```

<div id="dt_sklearn_snapml">
    <h2>Evaluate the Decision Tree Classifier Models</h2>
</div>


Run the the prediction to compute the probabilities of the test samples belonging to the class of fraudulent transactions using Decision Tree model


```python
y_pred_dt = dt_all.predict_proba(X_test)[:,1] 
```

Using these probabilities, evaluate the Area Under the Receiver Operating Characteristic Curve (ROC-AUC) score as a metric of model performance. 
The AUC-ROC score evaluates model's ability to distinguish positive and negative classes considering all possible probability thresholds. The higher its value, the better the model is considered for separating the two classes of values.
* Why column index 1 specifically?
For AUC-ROC score and most evaluation metrics, we want the probability of the positive class (fraud = 1), not the negative class. Column 0 would give the probability of not fraud which is the inverse and less useful.

```python
roc_auc_dt = roc_auc_score(y_test, y_pred_dt)
print('Decision Tree ROC-AUC score : {0:.3f}'.format(roc_auc_dt))
```

<div id="svm_sklearn_snap">
    <h2>Evaluate the Support Vector Machine Models</h2>
</div>


Run prediction to compute the probabilities of the test samples belonging to the class of fraudulent transactions using SVM model.

```python
y_pred_svm = svm_all.decision_function(X_test)
```

Evaluate the accuracy of SVM on the test set in terms of the ROC-AUC score.

```python
roc_auc_svm = roc_auc_score(y_test, y_pred_svm)
print("SVM ROC-AUC score: {0:.3f}".format(roc_auc_svm))
```

## Inferences and model assessment


- Currently using all 30 features of the dataset for training the models. Use the `corr()` function to find the top 6 features of the dataset to train the models on. 


```python
# Extract top 6 features of the dataset to train based on the correlation graph
correlation_values = abs(raw_data.corr()['Class']).drop('Class')
correlation_values = correlation_values.sort_values(ascending=False)[:6]
correlation_values
```

- Using only the above 6 features, modify the input variable for training.

```python
# Modify input for data training to use only top features [V3, V10, V12, V14, V16 and V17]
top_features = correlation_values.index.tolist()

# X: feature matrix 
#X = data_matrix[:, [3, 10, 12, 14, 16, 17]]
X_top6 = raw_data[top_features]

# y: labels vector
#y = data_matrix[:, 30]
y_top6 = raw_data["Class"]

# data normalisation
X_top6 = normalize(X_top6, norm="l1")

# print(raw_data.columns.to_list())
```

- Execute the Decision Tree model for this modified input variable and evaluate the change in ROC-AUC metric.

```python
# Resplit training data using data from top 6 features
X_train_top6, X_test_top6, y_train_top6, y_test_top6 = train_test_split(X_top6, y_top6, test_size=0.3, random_state=42)

# Recompute sample weight 
w_train_top6 = compute_sample_weight('balanced', y_train_top6)

print(X_train_top6.shape)
print(X_test_top6.shape)
```

```python
# Retrain Decision Tree model
dt_top6 = DecisionTreeClassifier(max_depth=4, random_state=42)
dt_top6.fit(X_train_top6, y_train_top6, sample_weight=w_train_top6)

# Evaluate
y_pred_top6_dt = dt_top6.predict_proba(X_test_top6)[:,1] 

roc_auc_top6_dt = roc_auc_score(y_test_top6, y_pred_top6_dt)
print('Decision Tree ROC-AUC (top6 features) score : {0:.3f}'.format(roc_auc_top6_dt))
```

- Execute the SVM model for this modified input variable and evaluate change in ROC-AUC metric.


```python
# Retrain SVM model
svm_top6 = LinearSVC(class_weight='balanced', random_state=42, loss="hinge", fit_intercept=False)
svm_top6.fit(X_train_top6, y_train_top6)

# Evaluate SVM model
y_pred_top6_svm = svm_top6.decision_function(X_test_top6)

roc_auc_top6_svm = roc_auc_score(y_test_top6, y_pred_top6_svm)
print("SVM ROC-AUC (top6 features) score: {0:.3f}".format(roc_auc_top6_svm))
```

Inferences on Decision Trees and SVMs:
- With a larger set of features, SVM performed relatively better in comparison to the Decision Trees.
- Decision Trees benefited from feature selection and performed better, suggesting the other features are adding noise.
- SVMs may require higher feature dimensionality to create an efficient decision hyperplane.


### Random Forest Classifier
- Random Forest is an ensemble method that builds multiple decision trees and combines their predictions. 
- It is expected to outperform a single Decision Tree by reducing variance through averaging.

```python
# Import libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
```

### Build and train RF classifier with HyperParameter Fine-Tuning 

```python
# Define parameter grid (use practice final project Course 9, module 6 parameters)
param_grid = {
    "n_estimators": [50, 100],
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5]
}

# Cross-validation method
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialise RF model and take into account class imbalance
rf = RandomForestClassifier(class_weight="balanced", random_state=42)

# Train model with HPT
rf_all_model = GridSearchCV(rf, param_grid=param_grid, cv=cv, scoring="roc_auc",
                           n_jobs=-1, verbose=1)
rf_all_model.fit(X_train, y_train)
print("Best parameters:", rf_all_model.best_params_)

# Model prediction
y_pred_rf = rf_all_model.best_estimator_.predict_proba(X_test)[:, 1]
roc_auc_rf = roc_auc_score(y_test, y_pred_rf)
print("RandomForestClassifier ROC-AUC score: {0:.3f}".format(roc_auc_rf))
```

```python
# Check data
print(X_train.shape)
print(X_test.shape)
```

```python
# Retrain RF model using the top 6 features
# Train model with HPT
rf_top6_model = GridSearchCV(rf, param_grid=param_grid, cv=cv, scoring="roc_auc",
                           n_jobs=-1, verbose=1)
rf_top6_model.fit(X_train_top6, y_train_top6)
print("Best parameters:", rf_top6_model.best_params_)

# Model prediction
y_pred_top6_rf = rf_top6_model.best_estimator_.predict_proba(X_test_top6)[:, 1]
roc_auc_top6_rf = roc_auc_score(y_test_top6, y_pred_top6_rf)
print("RandomForestClassifier ROC-AUC score (top6 features): {0:.3f}".format(roc_auc_top6_rf))
```

```python
# Map target names
fraud_outcome = {0: "Not fraud", 1: "Fraud"}
# print(classification_report(y_val, y_pred, target_names = list(fraud_outcome.values())))

# # Create confusion matrix to see the prediction
# print("\nConfusion Matrix:")
# print(confusion_matrix(y_val, y_pred))

# Generate classification reports
print("Decision Tree - All Features")
print(classification_report(y_test, y_pred_dt_all, target_names=list(fraud_outcome.values())))

print("Decision Tree - Top 6 Features")
print(classification_report(y_test_top6, y_pred_dt_top6, target_names=list(fraud_outcome.values())))

print("SVM - All Features")
print(classification_report(y_test, y_pred_svm_all, target_names=list(fraud_outcome.values())))

print("SVM - Top 6 Features")
print(classification_report(y_test_top6, y_pred_svm_top6, target_names=list(fraud_outcome.values())))

print("RandomForest - All Features")
print(classification_report(y_test, y_pred_rf_all, target_names=list(fraud_outcome.values())))

print("RandomForest - Top 6 Features")
print(classification_report(y_test_top6, y_pred_rf_top6, target_names=list(fraud_outcome.values())))
```

```python
# Visualise the confusion matrix
models = {"Decision Tree - All Features": (y_test, y_pred_dt_all),
          "SVM - All Features": (y_test, y_pred_svm_all),
          "RandomForest - All Features": (y_test, y_pred_rf_all),
          "Decision Tree - Top 6 Features": (y_test_top6, y_pred_dt_top6),
          "SVM - Top 6 Features": (y_test_top6, y_pred_svm_top6),
          "RandomForest - Top 6 Features": (y_test_top6, y_pred_rf_top6)
          }

# Plot layout
fig, axes = plt.subplots(2, 3, figsize = (12, 8))
axes = axes.flatten()

for i, (model_name, (y_true, y_pred)) in enumerate(models.items()):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=list(fraud_outcome.values()))
    disp.plot(cmap="RdPu", ax=axes[i], colorbar=False)
    axes[i].set_title(model_name, fontsize=11)
    
plt.suptitle("Confusion Matrices of All Models", fontsize = 14, y = 1.02)
plt.tight_layout()
plt.show()
    
```

### Model Comparison Inferences

- **Random Forest with all features is the best overall model** — it achieves 
  the highest fraud precision (0.71) and recall (0.83), making it most suitable 
  for real-world deployment where false positives have operational costs. 

- **Decision Tree and SVM prioritise recall over precision** — they catch most 
  fraud cases but generate large numbers of false positives, which would be 
  costly in production.

- **Feature reduction severely impacts SVM** — with top 6 features, SVM flags 
  36,401 legitimate transactions as fraudulent, making it operationally unusable. 
  The classification report confirms this: only 57% accuracy, meaning SVM with 6 
  features is flagging more than half of all legitimate transactions as fraudulent.
  This explains the dramatic ROC-AUC drop from 0.986 to 0.937.

- **ROC-AUC alone is misleading** — for example, SVM scored 0.986 ROC-AUC with all features 
  but has near-zero fraud precision, highlighting why multiple metrics are 
  essential for imbalanced datasets.

- **Random Forest is clearly the winner** - the only model with meaningful fraud
  precision(0.71 all features, 0.47 top 6). Every other model has near-zero fraud 
  precision, meaning they catch fraud but at the cost of high false positive cases.
